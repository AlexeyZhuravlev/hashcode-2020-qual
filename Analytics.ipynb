```python
import pandas as pd
import numpy as np
for filename in ['a', 'b', 'c', 'd', 'e', 'f']:
    with open('test/' + filename + '.in') as f:
        B, L, D = map(int, f.readline().strip().split())
        print('-'*30)
        print(filename)
        signup = [0 for i in range(L)]
        v = [0 for i in range(L)]
        n = [0 for i in range(L)]
        booksets = [set() for i in range(L)]
        
        print("Books:", B)
        print("Libraries:", L)
        print("Days:", D)
        
        scores = map(int, f.readline().strip().split())
        print("Sum books scores:", sum(scores))
        
        for i in range(L):
            n[i], signup[i], v[i] = map(int, f.readline().strip().split())
            booksets[i] = set(map(int, f.readline().strip().split()))
            n[i] = len(booksets[i])
        
        print("Mean number of books in library:", np.mean(n))
        print("Std number of books in library:", np.std(n))
        print()
        print("Mean signup:", np.mean(signup))
        print("Std signup: %.2f" % np.std(signup))
        print()
        print("Mean books per day:", np.mean(v))
        print("Std v: %.2f" % np.std(v))
        
        print()
        print("books overload", np.sum(n) / B)
            
        print()
```
